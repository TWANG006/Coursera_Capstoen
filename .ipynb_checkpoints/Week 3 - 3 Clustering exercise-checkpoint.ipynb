````python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

import numpy as np
from sklearn.cluster import KMeans
import matplotlib.cm as cm
import matplotlib.colors as colors
import folium

from geopy.geocoders import Nominatim
import requests
from bs4 import BeautifulSoup
````

### 1. Clean the data

````python
tn_df = df[df['Borough'].str.contains('Toronto')].reset_index(drop = True)
print(tn_df.shape)
tn_df.head()
````

````python
address = 'Toronto, Ontario'

geolocator = Nominatim(user_agent = 'toronto_explorer')
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print("The geographical coordinate of Toronto are [{}, {}].".format(latitude, longitude))
````

#### Create the map of Toronto and superimpose neighborhoods on top of it.

````python
# Generate the map centered at Toronto
map_toronto = folium.Map(location = [latitude, longitude], zoom_start = 10)

# Superimpose the neighborhoods 
for lat, lng, borough, neighborhood in zip(tn_df['Latitude'], tn_df['Longitude'], 
                                           tn_df['Borough'], tn_df['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html = True)
    folium.CircleMarker(
        location=[lat, lng],
        radius = 5, 
        popup = label,
        color = 'blue',
        fill = True,
        fill_color = '#33186cc',
        fill_opacity = 0.7,
        parse_html = False).add_to(map_toronto)
map_toronto    
````

#### Get all the venues in the neighborhoods in the ```tn_df``` data frame.

````python
def getNearbyVenues(names, latitudes, longitudes, limit = 100, radius = 500):
    venues_list = []
    
    for name, lat, lng in zip(names, latitudes, longitudes):
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            limit)
        
        # Meke the get queries
        results = requests.get(url).json()["response"]['groups'][0]['items']
        venues_list.append([(
            name,
            lat,
            lng,
            v['venue']['name'],
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])
        
        nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
        nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
        
    return (nearby_venues)
````

````python
tn_venues = getNearbyVenues(names = tn_df['Neighborhood'],
                            latitudes = tn_df['Latitude'],
                            longitudes = tn_df['Longitude'])

print(tn_venues.shape)
tn_venues.head()
````

### 3. Analyze Each Neighborhood

#### One-hot encoding on the Venue Category column

````python
# One hot encoding for the Venue Category
tn_onehot = pd.get_dummies(tn_venues[['Venue Category']], prefix="", prefix_sep = "")

# Add the Neighborhood column back
tn_onehot['Neighborhood'] = tn_venues['Neighborhood']

# Move the Neighborhood column to the first column
neighbor_col_id = tn_onehot.columns.get_loc('Neighborhood') # Id of the Neighborhood column
fixed_columns = [tn_onehot.columns[neighbor_col_id]] + list(tn_onehot.columns[:neighbor_col_id]) + list(tn_onehot.columns[neighbor_col_id + 1:-1])
tn_onehot = tn_onehot[fixed_columns]

print(tn_onehot.shape)
tn_onehot.head()
````

Check the mean number of categories for each neighborhood

````python
tn_grouped = tn_onehot.groupby('Neighborhood').mean().reset_index()
tn_grouped.head()
````

#### Find the n-th most common venues for each neighborhood

````python
def n_most_common_venues(row, n):
    row_cats = row.iloc[1:]
    row_cats_sorted = row_cats.sort_values(ascending = False)
    
    return row_cats_sorted.index.values[0:n]
````

Explore the __top 10__ common venue categories for each neighborhood

````python
n = 10

# Create the columns
indicators = ['st', 'nd', 'rd']

columns = ['Neighborhood']
for ind in np.arange(n):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# Create a new data frame
tn_neighborhood_venues_sorted = pd.DataFrame(columns = columns)
tn_neighborhood_venues_sorted['Neighborhood'] = tn_grouped['Neighborhood']

for ind in np.arange(tn_grouped.shape[0]):
    tn_neighborhood_venues_sorted.iloc[ind, 1:] = n_most_common_venues(tn_grouped.iloc[ind, :], n)

tn_neighborhood_venues_sorted.head()
````

### 4. Clustering

#### Run k-means to cluster the neighborhood into 6 clusters

````python
# The number of clusters
k = 10

# Drop the Neighborhood column
tn_grouped_clustering = tn_grouped.drop('Neighborhood', axis = 1)

print(tn_grouped_clustering.shape)

# Train the k-means
km_model = KMeans(n_clusters=k, random_state=0).fit(tn_grouped_clustering)

# Check clusters
km_model.labels_
````

Create a new data frame to combine the clusters and the top 10 venues

````python
# Insert the cluster label column
tn_neighborhood_venues_sorted.insert(0, 'Cluster Labels', km_model.labels_)

# merge the data frames
tn_merged = tn_df
tn_merged = tn_merged.join(tn_neighborhood_venues_sorted.set_index('Neighborhood'), on='Neighborhood')
print(tn_merged.shape)
tn_merged
````

Visualize the final results

````python
map_clusters = folium.Map(location=[latitude, longitude], zoom_start =10)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
color_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in color_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(tn_merged['Latitude'], tn_merged['Longitude'], tn_merged['Neighborhood'], tn_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
````
